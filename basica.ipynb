# Methods : 


##        instance          | class                | static

### 1. 1st parameter is self  | cls                | no compul. param
### 2. access the class and   | access class attri.   | none
###    instances atttributes
### 3. no decorator           | decorator             | decorator
###                            @classmethod            @static method

```python
class Laptop:
    storage_type ="ssd"

    def __init__(self,RAM,storage):
        self.RAM= RAM
        self.storage = storage

    @classmethod
    def get_storage_type(cls):
        print(f"storage_type = {cls.storage_type}")
        

    def get_info(self):
        print(f"laptop has {self.RAM} RAM and {self.storage} {self.storage_type}")

    @staticmethod
    def discount(price,discount):
        final_price = price - (discount*price/100)
        print(f"discounted price = {final_price}")

l1 = Laptop("16 gb","512 gb")
l1 = Laptop("8 gb","256 gb")

l1.discount(40_000,10)

l1.get_info()
l1.get_storage_type()
```

# ENCAPSULATION

wrapping data and functions in a single unit  ->> data hiding

data + methods = class

## data hiding 

public -     outside calss and inside class

protected -  inside class and subclasses   ## not enforced still accessible outside calss

private -    inside class only

```python
class BankAcc:
    def __init__(self,name,balance):
        self.__balance = balance  # private
        self._name = name         # protected
       

    def get_balance(self): # #getter
        return self.__balance

    def set_balance(self,newbalance): #setter
        self.__balance = newbalance

acc1 = BankAcc("prash",10_00_000)

acc1.set_balance(200_000) # use setter

print(acc1._name,acc1._BankAcc__balance) ## private attributes can be accessed in this way means theres no true private or protected its just convention

```

```python
# multiple Inheritance

class Teacher:

    def __init__(self,salary):
        self.salary = salary

class Student:
    def __init__(self,gpa):
        self.gpa =gpa

class TA(Teacher, Student):
    def __init__(self,salary,gpa,name):
        super().__init__(salary) # super se call kr rhe hai to self ki no need
        Student.__init__(self,gpa) # class name se call kr rhe hai to need
        self.name = name

ta1 = TA(15_000,8.5,"ant")

print(ta1.name,ta1.gpa,ta1.salary)
        
```

# ABSTRACTION 

        HIDING INTERNAL DETAIS AND SHOWING ONLY ESSENTIAL FEATURES

WE USE -  ABSTRACT CLASSES : BLUEPRINT FOR OTHER CLASSES (abc MODULE)

```python
from abc import ABC , abstractmethod

class Animal(ABC):
    @abstractmethod # decorator
    def make_sound():
        pass

        
class Lion(Animal):
    def make_sound(self):
        print("Roar!!")

class Cow(Animal):
    def make_sound(self):
        print("Mooo.....")

lion = Lion()
lion.make_sound()

cow = Cow()
cow.make_sound()
```

created make sound stuff in animal class but did not used it there rather we used in CHILD CLASSES

 # @abstract method :: 
                         pass - null value --> no work
                         parent class me no use but used in child classes
