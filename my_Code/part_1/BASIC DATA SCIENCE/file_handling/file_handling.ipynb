# Exception Handling

## try , except , else , finally




```python
try:
    x = int(input("enter value of x : "))
    ans = 10/x

except ZeroDivisionError:
    print("division by zero not allowed ")

except ValueError:
    print("division by string not allowed ")

else:# if any exception is raised this does not gets executed
    print("value after division: ", ans)


finally:
    print("end of the program *** ") # gets executed even  if there's an exception
```

# List Comprehensions

##  -->> reduces bulkiness of code

```python
# before :

squares = []

for i in range(6):
    squares.append(i*i)
print(squares)

# after :

sq = [i*i for i in range(6)]
print(sq)

# format :

## var = [output      iterable      condition]
```

```python
odd_sq = [i*i for i in range(6) if i%2!=0]
print(odd_sq)
```

```python
# replace negative no.s with zero : 

nums = [-2,-3,5,-4,-1,6,9]

nums = [0 if val<0 else val for val in nums]
print(nums)
```

```python
# conver to upper case :

words = ["hello", "this" , "dog","moo..."]

words = [val.upper() for val in words]
print(words)
```
