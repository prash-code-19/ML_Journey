```python
tips = sns.load_dataset("tips")
print(type(tips))
print(tips.head())
```

```python
# theme 

sns.set_theme(style="darkgrid")
sns.set_theme(style="ticks")
```

```python
#RELATIONAL PLOTS 
#1. SCATTER

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker", # decides color distincton like smoker v/s non-smokers
    size='size',
    style='smoker'
)

```

```python
# LINE  PLOT
xval=[1,2,3,4,5,6,7,8,9]
yval=[i**2 for i in xval]
sns.relplot(
    x=xval,
    y=yval,
    kind='line'
    
)
```

```python
# scatter

tips = sns.load_dataset('tips')

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
# LINE PLOT
#confidence interval(ci)-kitni probability hai  ki data shaded region me lie karega useally 95% ci +-5%
flights = sns.load_dataset("flights")

flights.head()
```

```python
sns.lineplot(
    data=flights,
    x='year',
    y="passengers"
)
```

```python
print(tips.head())
```

```python
sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"
)
```

```python
sns.boxplot(
    data=tips,
    x='day',
    y="tip"
)
```

```python
# distribution plot

penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
sns.histplot(
    data=penguins,
    x="body_mass_g",
    bins=30
)
```

```python
#matrix plot

# HEATMAPS --->>> IMPORTANT

#used to plot rectangular data as color-encoded-matrix

flights.head()

# show in every year how many passengers travelled each month
```

```python
# pivot the data

piv = flights.pivot(index="month", columns='year',values='passengers')
piv
```

```python
sns.heatmap(
    piv,
    cmap='coolwarm',
    annot=True,
    fmt="d"
)
plt.title("Passengers Heatmaps")
plt.tight_layout()
```

# BEST PRACTICES


usually  we will plot with Seaborn 

we will do customazisations using MatplotLib

```python
fig , ax  = plt.subplots()

sns.lineplot(
    data=tips,
    x="day",
    y='total_bill',
    hue="sex",
    marker="o",
    ax=ax,
    errorbar=None # hide the CI 
)

ax.set_title("avg bill value each day")
ax.set_xlabel("days")
ax.set_ylabel("Total Bill amount")
fig.tight_layout()

```
