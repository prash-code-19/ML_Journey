## Getting Started With Pandas

```python
import numpy as np
import pandas as pd
```

```python
# series - 1D labelled array
# dataframe - 2D labelled array

info = {
    "Name" : ["Rahul","Harshita","prash"],
    "CGPA" : [9.5,9.7,8.5]
}

df = pd.DataFrame(info)
df
```

***
## properties of Series
1. homogeneous - only one type of data is allowed, if entered an int in a grp of strings entire will be converted into object dataType
2. vectorised operations
3. Handle missing values with NAN
4. Mutable values , immutable values
***

```python
s= pd.Series([1,2,3,4,5])
print(s)
print(type(s))
print(s.dtype) # int

s= pd.Series([1,2,3,4,5,"hii"])
print(s.dtype) # object

#index

print(s[0])
print(s[2])
```

```python
# understanding Labels

age = pd.Series([45,23,16,33], index = ["adam","eve","bob","charlie"])
print(age)

print(age["adam"]) # labelled based access

print(age[0]) # index based access

# Access all index
print(age.index)
```

```python
# vectorized operations
s1=pd.Series([1,2,3,4,5])
s2 = pd.Series([5,6,7,8,9])

print(s1+s2) 

# Mutable Values

s1[0] = 100
print(s1)

# Immutable Size

#s1.drop(1) # no change is s1
#print(s1)
changed_s1 = s1.drop(0)

print(s1)

print(changed_s1)
```

```python
# creating dataframe using DICTIONARY

info = {
    "Name" : ["adam","eve","wai"],
    "Age" : [23,24,35],
    "GPA" : [9.5,8.4,7.8]
}

df = pd.DataFrame(info)
print(df , type(df))

print(df.index)#row values (prints nums in range if index not explicitly defined)
print(df.columns)
```

```python
# creating Dataframe using LIST of LISTS

df = pd.DataFrame([["Adam",23],["eve",24],["wai",35]], columns = ["Name","age"])
df
```

```python
# creating DataFrame using Numpy Arrays

np_arr= np.array([[1,2,3],[4,5,6]])
df = pd.DataFrame(np_arr,columns = ['a','b','c'])
df
```
