## query() Rules
1. expressions as strings
2. Column based reference
3. Backtricks for col_name with space or special chars
4. Operators (&,|,~,>=,<=,<,>,== etc.)
5. @ to refrence Python vars
6. chanined comparisons
***

```python
# Query () - returns a copy

df.query("aqi > 100 & timestamp == '2025-11-04 18:25:17.554219'")

my_country = "IN"
df.query("country == @my_country") # we can use @ to directly use variables

df["timestamp"] = pd.to_datetime(df["timestamp"])
df["date"] = df["timestamp"].dt.date
selected_date = pd.to_datetime("2025-11-04").date()

df.query("country == @my_country & date == @selected_date & aqi > 100")

# Use .copy() to avoid confusion
countries = df[["country", "aqi"]].copy()
countries.loc[0:4, "country"] = "India"

print(countries)
```

# Feature Engineering

## Transforming Data (existing  -> new)
1. apply()
2. map()
3. assing()
4. replace(old,new)
5. rename(columns={old,new})
6. sort_values()

```python
# transforming data

df2 = df.copy()

df2["tax"] = df2["income"].apply(lambda x: "20%" if x>=50000 else "10%")

gender_map ={
    "Male":"M",
    "female":"F",
    "Unknown":"U"
}

df2["gender"] = df2["gender"].map(gender_map)


df2.assign(new_income = df["income"]*1.1)


df2["country"] = df2["country"].replace("USA","US")

df2
```

```python
# Transformation - Sorting
df2 = df.copy()

# Renaming
df2.columns = ["Id", "Name", "Age", "Country", "Gender", "Income"]
df2.rename(columns={"Income": "Salary"})
df2.rename(index={0: "first"})

# Sorting - values & index
df2.sort_values("Income")                     # sort values in ascending
df2.sort_values("Income", ascending=False)    # sort values in descending
df2.sort_values(["Age", "Income"])            # sorts for age, if age is same then sorts for income 

sorted_df2 = df2.sort_values(["Age", "Income"])
sorted_df2.sort_index()

# Reset Index
sorted_df2.reset_index()
sorted_df2.reset_index(drop=True)     # to drop original index vals

# Ranking
df2["Ranking"] = df2["Income"].rank(ascending=False, method="dense")

# Reorder
df2[["Id", "Name", "Age", "Income", "Gender", "Country", "Ranking"]]
```

# Melt and Pivot

1. Melt - wide format to long format
2. Pivot - long ot wide

1. melt(id_vars(KEEP),value_vars(UNPIVOT),var_name(NEW_COL_NAME),value_name(NEW_COL_VALUE))
2. pivot(index(KEEP),columns,values)

***

```python
df = pd.DataFrame({
    "country":["USA","USA","India","India"],
    "year": [2020,2021,2020,2021],
    "sales":[100,120,90,110],
    "profit":[20,25,18,22]
})

df
```

```python
melted_df=df.melt(
    id_vars=["country","year"],
    value_vars=["sales","profit"],
    var_name="metrics",
    value_name="value"
)

melted_df
```

```python
original = melted_df.pivot(
    index = ["country","year"], # unique value
    columns = "metrics",
    values= "value"
)

original
```

# merging and Joining
1. inner join
2. left join
3. right join
4. outer join
***

```python
df_customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "name": ["Adam", "Bob", "Charlie", "Dave"]
})

df_orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104],
    "customer_id": [2, 1, 4, 5],
    "amount": [250, 120, 300, 180]
})
```

```python
df_customers
```

```python
df_orders
```

```python
pd.merge(df_customers,df_orders,on="customer_id") # inner join - common in both the table like in this case thecustomer_id
```

```python
pd.merge(df_customers,df_orders,on = "customer_id",how="left") # left join - keep all the data from left table and replace values in other table with nan if not available
```

```python
pd.merge(df_customers,df_orders,on = "customer_id",how="right")
```

```python
pd.merge(df_customers,df_orders,on = "customer_id",how="outer") #if values not available then replace with NAN
```

# Data Concatenation

stacking DataFrames on the top of each other(rows) or side by side(columns)
***

```python
# Concatenation
df1 = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Adam", "Bob", "Charlie"]
})

df2 = pd.DataFrame({
    "id": [4, 5, 6],
    "name": ["David", "Eva", "Frank"]
})

pd.concat([df1, df2])                       # Row wise concatenation
pd.concat([df1, df2], axis=1)               # Col wise concatenation
pd.concat([df1, df2], ignore_index=True)    # Row wise concatenation - new index
```
