```python
url = "https://www.scrapethissite.com/pages/simple/"
res = requests.get(url)
```

## Status codes

1. 200 - OK
2. 301/302 - REDIRECT
3. 400 - BAD REQUEST
4. 401 - UNAUTHORISED
5. 403 - FORBIDDEN
6. 404 - NOT FOUND
7. 500 - SERVER ERROR

```python
if res.status_code == 200:
    print(res.text) # HTML CONTENT OF THE PAGE AS A STRING
    print("\n \n**************************************\n")
    print(res.content) # RAW BYTES OF THE RESPONSE
    print("\n \n**************************************\n")
    print(res.headers) #  META DATA LIKE CONTENT TYPE, SERVER INFO ETC.
```

### DOWNLOADING AND STORING THESE FILES

```python

with open("scrapped_data/data1.html","w", encoding="utf8") as f:
    f.write(res.text)
```

```python
from bs4 import BeautifulSoup

with open("scrapped_data/data1.html","r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content,"lxml")
```

```python
#soup
```

## METHODS IN SOUP

1. find() : finds the first element
2. find_all() : finds all the elements of the tag
3. find_parent() : finds the parent of a particular tag
4. find_next() : travel forward (next sibling)
6. find_previous() : travels backward

## others :

### css selectors :
1. select()
2. select_one()

### attributes :
1. get()
2. .attrs



```python
soup.find("h1")
```

```python
all_h3 = soup.find_all("h3")

all_countries =  []

for h3 in all_h3:
    name = h3.get_text(strip = True) # .get_text() - to access the content inside h3
    #print(h3.find_parent("div")) # to find the div parent
    #print(h3.find_parent("div")["class"]) # to find the class of the div parent
    #print(h3.find_next("div")) # finds the next div element after the previous found tag
    population = h3.find_next("div").select("span.country-population")[0].get_text(strip=True)
    print(h3.find_next("div").select_one("span.country-population").get_text(strip=True))

    all_countries.append([name,population])
    
    
```

```python
print(all_countries)
```

```python
df=pd.DataFrame(all_countries,columns=["Nmae","population"])
df
```

```python
df.to_csv("scrapped_data/data.csv",index=False)
```
