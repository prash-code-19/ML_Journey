```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
ass_data = pd.read_csv("HousePricePrediction.csv")
```

```python
ass_data
```

```python
ass_data.isnull().sum()
```

```python
ass_data.duplicated().sample(15)
```

```python
ass_data.drop(["Id"], axis=1,inplace=True)

# replacing null values of sales with mean of sales

ass_data["SalePrice"]=ass_data["SalePrice"].fillna(ass_data["SalePrice"].mean())

# drop records with null values

ass_data = ass_data.dropna()
```

```python
# Encoding data

cols = ['MSZoning', 'LotConfig', 'BldgType', 'Exterior1st']
ass_data = pd.get_dummies(ass_data,columns=cols,drop_first=True)
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split

X = ass_data.drop(["SalePrice"],axis=1)
y=ass_data["SalePrice"]

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.head()
```

```python
#Train Linear Regression Model

from sklearn.linear_model import LinearRegression as LR

model=LR()
model.fit(X_train,y_train)
```

```python
# Model Evaluation
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
import numpy as np

y_pred = model.predict(X_test)

print("R2 Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print(mean_absolute_percentage_error(y_test, y_pred))
```
