```python
import json
```

```python
# load the data

def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data

```

```python
data = load_data("data.json")

print(data)
```

# cleaning and structuring :

1. data is inconsistent (mixed data - int,float,str)
2. missing data (age)
3. duplicate (alice is )

```python
def clean_data(data):
    text_to_num = {"one":1,"two":2,"three":3,"four":4,"five":5}

    cleaned_data = []
    unique_users = set()


    for user in data:
        #clean ratings

        raw_rating = str(user["rating"]).strip().lower() # strip method clears spaces
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        print(raw_rating)

        user["rating"] = raw_rating

        # handle missing value

        raw_age = user.get("age") # get() returns None in case of missing key
        if(raw_age == None):
            user["age"] = None

        # De-duplication:
        if(user["name"].strip() in unique_users):
            continue
        unique_users.add(user["name"])
        cleaned_data.append(user)
    return cleaned_data

        

            
clean_data(data)
```

## assignment_2
## getting meaningful insights

```python
data = clean_data(data)
```

```python
def get_insights(data):

    #avg rating

    tot_rating = 0
    for user in data:
        tot_rating += float(user["rating"])

    print(f"avg rating = {tot_rating/len(data)}")


    # percentage of users with poor rating

    poor_rating = 0
    for user in data:
        if(float(user["rating"])<3):
            poor_rating +=1

    print(f"% of user with poor rating = {poor_rating/len(data)*100}%")
```

```python
get_insights(data)
```

# recommendation features

## # if user_rating >= 4 -->> recommend same brand
## # if user_rating < 4 -->> recommend other brand

```python
# build recommendation feature

def get_recom(data):
    recommendation = []
    

    for user in data:
        curr_recom ={}
        curr_recom["name"] = user["name"]

        if(float(user["rating"]) >=4):
            curr_recom["brand"]= "Apple"

        else:
            curr_recom["brand"] = "Samsung"

        recommendation.append(curr_recom)
    return recommendation
            
```

```python
get_recom(data)
```
