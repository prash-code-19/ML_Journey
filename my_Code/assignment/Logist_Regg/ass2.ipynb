```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
emp = pd.read_csv("employee_turnover.csv")
```

```python
emp.head()
```

```python
X = emp.drop("Employee_Turnover",axis=1)
y = emp['Employee_Turnover']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,y, test_size = 0.25 , random_state= 42
)
```

```python
model = LogisticRegression(max_iter=200)
model.fit(X_train,y_train)
```

```python
# L1 Regularization (Lasso)

lasso = LogisticRegression(penalty='l1',solver='liblinear',C=0.5)

lasso.fit(X_train,y_train)
```

```python
# L2 Regularization (Ridge)

ridge  =LogisticRegression(penalty="l2",C=1,max_iter=200)
ridge.fit(X_train,y_train)
```

```python
# Model Evaluatoin

models = {"Baseline":model , "Lasso":lasso , "Ridge":ridge}

for name,x in models.items():
    y_pred = x.predict(X_test)

    print(f"\n{name}")
    print("Accuracy : ", accuracy_score(y_test,y_pred))
    print(classification_report(y_test,y_pred
                               ))
```
