```python
import numpy as np
```

```python
# Logisstic Regression

class LogisticRegression:

    def  __init__(self, learning_rate = 0.1, epochs = 1000):
        self.bias = None
        self.weights = None
        self.lr = learning_rate
        self.epochs = epochs


    def _sigmoid(self,z):
        return(1/ (1+np.exp(-z)))


    def fit(self , X, y):
        m,n = X.shape

        self.bias = 0
        self.weights = np.zeros(n)

        for i in range(self.epochs):
            z=self.bias + np.dot(X, self.weights)
            y_pred = self._sigmoid(z)

            db = (1/m)*(y_pred - y)
            dw = (1/m)*np.dot(X.T,(y_pred - y))

            self.bias -= self.lr*db
            self.weights -= self.lr*dw

    # to do predictions first we calculate prediction values and then convert those predictions into true/false

    def get_probabilities(self, X):
        z = self.bias + np.dot(X, self.weights)
        return self._sigmoid(z) # will return value between 0 and 1

    def predict(self, X, threshold=0.5):
        probabilities = self.get_probabilities(X)
        y_pred_bool = probabilities>= threshold # returns array of true and false
        return y_pred_bool.astype(int) #convert boolean into 0 and 1

            
```

```python
X = np.array([[1,2],[2,3],[3,4],[4,5]])
y = np.array([0,0,1,1])

model = LogisticRegression()
model.fit(X,y)

y_pred = model.predict(X)
print(y_pred)

print(model.get_probabilities(X))
```
