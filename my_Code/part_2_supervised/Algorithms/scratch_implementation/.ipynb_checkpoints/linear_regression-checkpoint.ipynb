```python
# Linear Regression

class LinearRegression:

    def __init__(self, learning_rate = 0.01, epochs = 1000):
        
        self.bias = None # initializeed with a none value
        self.weights = None
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self, X_train, y_train): # X_train, y_train

        m,n = X_train.shape # (number of samples , number of features)

        # Step 1:
        self.bias = 0 # only one bias term
        self.weights = np.zeros(n) # theta1, theta2, ... theta n initializing all of the thetas(features) woth zero

        # Gradient Decent
        for i in range(self.epochs):
            # Step 2:
            y_pred = self.bias + np.dot(X_train , self.weights)
    
            
            # Step 3: calculating deratives
            db = (1/m) * np.sum(y_pred - y_train) # diff of bias
            dw = (1/m) * np.dot(X_train.T, (y_pred - y_train)) # diffr of weights
    
    
            # Step 4: update values
            self.bias -= self.lr*db
            self.weights -= self.lr * dw
        
        
    def predict(self, X_test):
        y_pred = self.bias + np.dot(X_test, self.weights)
        return y_pred
        
```

```python
X = np.array([[1],[2],[3],[4],[5]])
y = np.array([2,4,6,8,10])

model = LinearRegression()
model.fit(X,y)

y_pred = model.predict(X)
print(y_pred)
```

```python
# OLS = Ordinary Least Squares

# Normal equation is used  and since it is closed form it directly gives output

class LinearRegressionOLS:

    def  __init__(self):

        self.bias = None
        self.weights = None

    def fit(self,X,y):
        m,n = X.shape

        X_updated = np.c_[np.ones((m,1)),X] # added 1's in forst column to adjust theta not

        #normal equation

        #theta = np.dot(np.linalg.inv(np.dot(X_updated.T , X_updated)), np.dot(X_updated.T,y))
        theta = np.linalg.inv(X_updated.T @ X_updated) @ X_updated.T @ y

        self.bias = theta[0]
        self.weights = theta[1:]

    def predict(self,X):
        y_pred = self.bias + np.dot(X,self.weights)
        return y_pred        
```

```python
X = np.array([[1],[2],[3],[4],[5]])
y = np.array([2,4,6,8,10])

model = LinearRegressionOLS()
model.fit(X,y)

y_pred = model.predict(X)
print(y_pred)
```
