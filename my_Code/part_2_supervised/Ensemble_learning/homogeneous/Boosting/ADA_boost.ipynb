```python
from sklearn.datasets import make_classification , make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, r2_score as r2
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
```

```python
# make dataset

X, y = make_classification(
    n_samples = 500,
    n_features = 20,
    n_informative = 10,
    n_redundant = 2,
    random_state = 42
)

X_train , X_test , y_train, y_test = train_test_split(
    X , y, test_size = 0.3, random_state =42
)
```

```python
# make a DT stump - Base Model

base_model = DecisionTreeClassifier(
    max_depth = 1
)
```

```python
# ADA BOOST classifier

from sklearn.ensemble import AdaBoostClassifier

abc = AdaBoostClassifier(
    estimator = base_model,
    n_estimators = 250,# no of times base models u wish to use
    random_state = 42
)
```

```python
abc.fit(X_train,y_train)
```

```python
y_pred = abc.predict(X_test)
print("accuracy : ", accuracy_score(y_test,y_pred))
```

```python
print("calssification report", classification_report(y_test, y_pred))
```
